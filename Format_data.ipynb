```python
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies=pd.read_csv('movies.csv')
ratings=pd.read_csv('ratings.csv')
movies.head()
```

```python
#Specify the parantheses so they won't conflict with movies that have years in their titles
movies['year']=movies.title.str.extract('(\(\d\d\d\d\))', expand=False)

#Remove the parantheses
movies['year']=movies.year.str.extract('(\d\d\d\d)', expand=False)
```

```python
#Remove years from 'title' column
movies['title']=movies.title.str.replace('(\(\d\d\d\d\))','')

#Remove possible ending whitespaces 
movies['title']=movies['title'].apply(lambda x: x.strip())
movies.head()
```

```python
#Split values in Genres into a list of Genres to simplify
movies['genres']=movies.genres.str.split('|')
movies.head()
```

Apply One Hot Encoding to convert the list of genres to a vector where each column corresponds to the value

```python
#copy the movies dataset into a new dataset since this won't be used in the first case
genres_df=movies.copy()

#iterate through the list of genres and place 1 into the corresponding column
for index, row in movies.iterrows():
    for genre in row['genres']:
        genres_df.at[index,genre]=1

genres_df=genres_df.fillna(0)
genres_df.head()
```
